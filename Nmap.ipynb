**Nmap**(Network Mapper) is a security scanner used to discover hosts and services on a computer network

Use this command to install: *pip install python-nmap*

```python
import nmap
myscan = nmap.PortScanner()
myscan.scan('192.168.100.176', '21-443')
```

```python
myscan.scaninfo()
```

```python
myscan['192.168.100.176'].state()
```
